### Plot Pore size distribution of 7 loaded IPA flexible UiO66 and Rigid UiO-66

```python
import numpy as np
import ase
from ase import Atoms
from ase.io import read, write
from collections import Counter
import matplotlib.pyplot as plt
```

```python

def Plot_PSD(filename,labels,rigid_file):
    big_dia=[]
    big_PSD=[]
    for file in filename: 
        small_dia=[]
        small_PSD=[]
        for i in range(0,10):
            dia,connoly,PSD=np.loadtxt(f'{file}/{snaps[i]}/PoreSizeDistributionHistogram/System_0/HistogramPoreSizeDistribution_finalmovie-{snaps[i]}_1.1.1_300.000000_0.000000.dat',unpack=True)
            small_dia.append(dia)
            small_PSD.append(PSD)
        print(len(big_PSD))
        big_dia.append(small_dia)
        big_PSD.append(small_PSD)
        
    Plot(big_dia,big_PSD,labels,rigid_file)
    
def Plot(big_dia,big_PSD,labels,f3): 
    plt.figure(figsize=(15,10))
   
    for i in range(len(big_dia)):
        
        plt.plot(np.average(big_dia[i],axis=0),np.average(big_PSD[i],axis=0),'-',label=labels[i],linewidth=2);
        
        plt.xlim(0,8.5)
        plt.legend(frameon=False, prop = {'size':25},fontsize='x-large')
        plt.tick_params(direction = 'in', right = True, top = True)
        plt.tick_params(axis='both',labelsize = 25,length=10, width=2)
        plt.xlabel(r'Pore diameter ($\rm \AA$)', size = 25)
        plt.ylabel('Pore size distribution', size = 25);
        
    plt.plot(f3[:,0],f3[:,2],'m-',alpha=0.4,label='Rigid')
    plt.xlim(0,8.5)
    plt.legend(frameon=False, prop = {'size':25},fontsize='x-large')
    plt.tick_params(direction = 'in', right = True, top = True)
    plt.tick_params(axis='both',labelsize = 25,length=10, width=2)
    plt.xlabel(r'Pore diameter ($\rm \AA$)', size = 25)
    plt.ylabel('Pore size distribution', size = 25);
    plt.savefig('/ihome/kjohnson/pbs13/FlexibleFramework-resultanalysis/PoreSizeDistribution.png',dpi=400, figsize = (10,10), bbox_inches = 'tight')




    
        
```

```python
Path_7loadedIPa='/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/7loadedIPA/AllMovies'
Path_7loadedAcetone='/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/7loadedAcetone/AllMovies'
Path_23N2='/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/23loadedN2/AllMovies'
Path_empty298K='/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/empty-298K/AllMovies'
Path_77K='/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/empty-77K/AllMovies'
Path_RigidUiO66='/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/RigidUiO66/PoreSizeDistributionHistogram/System_0'
snaps=['3','5','7','9','12','15','17','18','20','21']
f3=np.loadtxt('/bgfs/kjohnson/pbs13/FlexibleMOFs-GCMC/PoreSizeDistribution/RigidUiO66/PoreSizeDistributionHistogram/System_0/HistogramPoreSizeDistribution_final_1.1.1_300.000000_0.000000.dat')

```

```python
files=[Path_7loadedIPa,Path_7loadedAcetone]
identity=['7 loaded IPA (Flexible)','7 loaded acetone (Flexible)']
Plot_PSD(files,identity,f3)

```
